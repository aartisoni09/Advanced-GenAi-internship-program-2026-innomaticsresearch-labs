```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
# highest score
highest_score = max(employees.values())
# all employees having highest score
top_performers = [name for name, score in employees.items() if score == highest_score]
# output
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

```python
query = "Buy mobile phone buy phone online"

# convert to lowercase
query = query.lower()
# remove punctuation
import string
for ch in string.punctuation:
    query = query.replace(ch, "")

# split words
words = query.split()
# count frequency
freq = {}
for word in words:
    freq[word] = freq.get(word, 0) + 1
# keep only words searched more than once
result = {word: count for word, count in freq.items() if count > 1}

print(result)
```

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for i, value in enumerate(sensor_readings):
    if value % 2 == 0:   # only even readings
        valid_readings.append((i, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# count domains
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

total = len(emails)

# print percentage usage
for domain, count in domain_count.items():
    percentage = (count / total) * 100
    print(f"{domain}: {int(percentage)}%")
```

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# calculate average sales
average = sum(sales) / len(sales)

# 30% above average threshold
threshold = average * 1.3

# detect spike days
for i, value in enumerate(sales):
    if value > threshold:
        print(f"Day {i+1}: {value}")
```

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

count = {}

# count each user id
for uid in user_ids:
    count[uid] = count.get(uid, 0) + 1

# print only duplicates
for uid, times in count.items():
    if times > 1:
        print(f"{uid} → {times} times")
```
