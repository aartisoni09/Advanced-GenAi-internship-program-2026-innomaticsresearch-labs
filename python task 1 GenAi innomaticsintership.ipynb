```python
# Given marks
marks = [45, 78, 90, 33, 60]
pass_count = 0
fail_count = 0

# Checking result
for m in marks:
    if m >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Final result
print("Total students:", len(marks))
print("Total pass students:", pass_count)
print("Total fail students:", fail_count)
```

```python
# Given list of names
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = []
#forloop
for name in names:
    clean_name = name.strip().lower()
    cleaned_names.append(clean_name)
# Print cleaned list
print(cleaned_names)
#list and list funtions are used
```
