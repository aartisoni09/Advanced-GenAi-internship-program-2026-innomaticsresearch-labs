```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: "))

        if amount < 50:
            print("Recharge amount must be at least ₹50. Try again.")
        elif amount in valid_plans:
            print("Recharge successful with plan ₹", amount)
            break
        else:
            print("Invalid plan. Available plans are:", valid_plans)

validate_recharge()
```

```python
def check_inventory(products):

    for item, stock in products.items():

        if stock < 15:
            print(item, ": Reorder Alert")
        else:
            print(item, ": Stock OK")


products = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 8,
    "Monitor": 20
}

check_inventory(products)
```

```python
def student_result(marks):

    total = 0

    for m in marks:
        total += m

    average = total / len(marks)

    if average >= 50:
        result = "Pass"
    else:
        result = "Fail"

    return average, result


marks = [60, 55, 40, 70, 65]

avg, res = student_result(marks)

print("Average Marks:", avg)
print("Result:", res)
```

```python
def calculate_fare(distance, is_peak_hour):
    base_fare = 50
    rate_per_km = 12
    
    total_fare = base_fare + (distance * rate_per_km)
    
    if is_peak_hour == "yes":
        total_fare *= 1.25  # 25% extra for peak hour
        
    return total_fare

while True:
    dist = float(input("\nEnter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ").lower()
    
    fare = calculate_fare(dist, peak)
    print(f"Your total fare is: ₹{fare:.2f}")
    
    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        print("Thank you for using the Cab Fare Estimator!")
        break
```
