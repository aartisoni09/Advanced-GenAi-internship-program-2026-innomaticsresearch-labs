```python
# Input sentence
sentence = "python is easy and python is powerful"

# Split sentence into words
words = sentence.split()

# Convert list to set to get unique words
unique_words = set(words)

# Print results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
# Given dictionary of employees or salaries
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# employee with highest salary
name, salary = max(employees.items(), key=lambda x: x[1])

print(f"Highest Salary: {name} - {salary}")
```

```python
# Given list of numbers
numbers = [45, 22, 89, 10, 66]
#  max and min with first element
maximum = numbers[0]
minimum = numbers[0]

# for loop
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num
#print numbers
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

```python
prices = [450, 1200, 899, 1500, 300]

# Initialize count with 0
count = 0
for price in prices:
    if price > 1000:
        count += 1

# Print result
print("Products above 1000:", count)
```

```python
#list
attendance = ["P", "P", "A", "P", "P"]
present_count = attendance.count("P")
# Total days
total_days = len(attendance)

# Calculate attendance percentage
attendance_percentage = (present_count / total_days) * 100

print("Attendance Percentage:", attendance_percentage)
```

```python
#  string
text = "pythonp"

# Dictionary to store
char_count = {}
# Count each character
for ch in text:
    if ch in char_count:
        char_count[ch] += 1
    else:
        char_count[ch] = 1

print(char_count)
```
